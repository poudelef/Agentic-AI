```python
# Create API Client
from anthropic import Anthropic

client = Anthropic()
model = "claude-opus-4-1-20250805"

# Load environment variables from .env file
from dotenv import load_dotenv

load_dotenv()


```

```python
def add_user_message(messages, text):
    user_message= {
        "role": "user",
        "content": text
    }
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {
        "role": "assistant",
        "content": text
    }
    messages.append(assistant_message)   

def chat(messages, system = None):   
    prompt = {
        "model":model,
        "max_tokens":1000,
        "messages": messages,
    } 

    # We Can't pass system promot of none
    if system:
        prompt["system"]= system
    
    messages = client.messages.create(**prompt)
    
    return messages.content[0].text
```

```python
system = """
    You are a math tutor. You will help solving the answer by 
    showing the way or doing similar problem but you don' give the answer that easily.
    
    """ 

messages = []
add_user_message(messages, "How to solve 5x + 2 = 12")
chat(messages,system=system)

```

```python
system = """
You are a computer programer and you generate code that is clean and easy to understand and effective. 
You will take space new line and wrtie code that is very clean with less symboles.
"""
messages = []
add_user_message(messages, "Wrtie a python function that checks a string for duplicate characters.")
chat(messages, system)
```
