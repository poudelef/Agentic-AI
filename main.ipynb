```python
def add_user_message(messages, text):
    user_message= {
        "role": "user",
        "content": text
    }
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {
        "role": "assistant",
        "content": text
    }
    messages.append(assistant_message)   

def chat(messages):     
    messages = client.messages.create(
        model=model,
        max_tokens=1000,
        messages= messages
        
    )
    return messages.content[0].text
```

```python
# Make string list of messages
messages = []

# Add the initial user question of "Define quantum computing" in one sentence
add_user_message(messages,"Define Quantum Computing in one sentence")

# Pass the list of messages into 'chat' to get an answer
answer = chat(messages)
# answer

# Take the answer and add it as an assistant messages into our list
add_assistant_message(messages, answer) 

# Add in the user's follow-up question
add_user_message(messages,"Give me one more phrase")

# call chat again
answer = chat(messages)

add_assistant_message(messages, answer)
```
